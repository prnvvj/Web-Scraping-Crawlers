```python
# Creating a function that create month while given input as date ----------
def month_converter(date):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    M_number=months.index(str(date.split('/')[0][:3])) + 1  
    return str(M_number)+"/"+str(date.split('/')[1])+"/"+str(date.split('/')[2])
```

```python
# Reading the URLs from a csv file ----------
dataCsv = pd.read_csv(r"D:\WFH\WebScraping\Amazon Monthly URLs.xlsx - Sheet1.csv")
```

```python
main_list=[]
for i in tqdm(dataCsv['URL'].values[:10]):
    try:
        driver= webdriver.Chrome("D:/WFH/WebScraping/chromedriver.exe")
        driver.get(i.strip())
        driver.maximize_window()
        sleep(5)
        page = driver.page_source
        sel=Selector(text=page)
        url=i
        asin=i.split('/')[-1]
        name=str(sel.xpath("//span[@id='productTitle']/text()").get()).replace('\n','').strip()

        overall_start_rateing=sel.xpath('//span[@class="a-size-medium a-color-base" and contains(text(),"out of")]/text()').get()

        totle_review=re.sub('[^0-9]',"",str(sel.xpath('//span[@class="a-size-base a-color-secondary" and contains(text(),"global ratings")]/text()').get()).strip())

        review_title=sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//a[@data-hook="review-title"]/span/text()').get()

        StarRating=str(sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//a[@class="a-link-normal"]/@title').get()).split()[0]

        reviwer_name=sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//div[@class="a-profile-content"]/span/text()').get()

        review_date=sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//span[@data-hook="review-date"]/text()').get()

        #-------------------
        review_date_f=month_converter(str(review_date.split('on')[1].strip().replace(',','').replace(' ','/')))
        #-------------------

        review_description=str(sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//div[@data-hook="review-collapsed"]/span/text()').get()).strip()

        other_tag1=sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//span[@data-hook="avp-badge-linkless"]/text()').get()

        size_bought=sel.xpath('(//div[@class="card-padding"]//div[@class="a-section review aok-relative"])[position()=1]//span[@data-hook="format-strip-linkless"]/text()').get()    

        M_all=[url,asin,name,overall_start_rateing,totle_review,review_title,StarRating,reviwer_name,review_date_f,review_description,other_tag1,size_bought]

        #print(M_all)

        main_list.append(M_all)

        driver.close()
    except Exception as e:
        print(e)
        continue
```
